```

import math
import scipy.stats as st
import pandas as pd
import numpy as np
```

```
# 1 

n = 30
vals = 3

ssr = 13.2361
sst = 20.4825

sse = sst - ssr

print(f"SSR: {ssr}")
print(f"SSE: {sse}")
print(f"SST: {sst}")

df_reg = vals - 1
df_err = n - vals
df_tot = n - 1

print(f"df reg: {df_reg}")
print(f"df err: {df_err}")
print(f"df tot: {df_tot}")

msr = round(ssr / df_reg, 4)
mse = round(sse / df_err, 4)

print(f"MSR: {msr}")
print(f"MSE: {mse}")

f = msr / mse
print(f"F-val: {f}")

p_val = round(st.f.cdf(f, df_reg, df_err), 6)
print(f"P val: {1 - p_val}")


alpha = 0.05

f_crit = st.f.ppf(1-alpha, df_reg, df_err)
print(f"Crit val: {f_crit}")
```

```
# 2 - chi square

chi_sq_calc = 25.12

k = 3
df_k = k - 1
alpha = 0.01

crit_val = round(st.chi2.ppf(1 - alpha, df_k), 4)
print(f"Critical value: {crit_val}")

claim = "that the proportions are not equally dispersed over the five possible responses"

if chi_sq_calc > crit_val:
    print(f"At alpha = {alpha}, there is sufficient evidence to state {claim}")
else:
    print(f"At alpha = {alpha}, there is not sufficient evidence to state {claim}")
```

```
# 4 - chi-square test of homogeneity 

val_setup = {
    'units' : ['0-2500', '2501-8000', '8001-15000', '15001-20000'],
    'expected': [11, 23, 56, 15],
    'observed': [0, 20, 83, 52]
}

vals = pd.DataFrame(val_setup)

vals.set_index('units', inplace=True)
vals
```

```
k = len(vals)
df_k = k - 1
alpha = 0.05

crit_val = round(st.chi2.ppf(1 - alpha, df_k), 4)
print(f"Critical value: {crit_val}")

chi_sq_calc = sum( ( (vals.observed.values - vals.expected.values)**2) / vals.expected.values)
print(f"Test statistic: {round(chi_sq_calc, 4)}")

claim = "there a change in the historical proportions of defective motocycle components"

if chi_sq_calc > crit_val:
    print(f"At alpha = {alpha}, there is sufficient evidence to state {claim}")
else:
    print(f"At alpha = {alpha}, there is not sufficient evidence to state {claim}")
```

```
# 5 - randomized block

t = 3
print(f"t: {t}")

b = 5
print(f"b: {b}")

N =  b * t
print(f"N: {N}\n")

df_b = b - 1
df_t = t - 1
df_e = (t - 1) * (b - 1)
df_N = N - 1

print(f"DFT: {df_t}\nDFB: {df_b}\nDFE: {df_e}\nDFN: {df_N}\n")

tss = 624536.4

sst = 228679.6
print(f"SST: {sst}")

ssb = 125483.1
print(f"SSB: {ssb}")

sse = round(tss - sst - ssb, 4)
print(f"SSE: {sse}")
print(f"TSS: {tss}\n")

mst = round(sst / df_t, 4)
print(f"MST: {mst}")

msb = round(ssb / df_b, 4)
print(f"MSB: {msb}")

mse = round(sse / df_e, 4)
print(f"MSE: {mse}\n")

F_t = round(mst / mse, 4)
F_b = round(msb / mse, 4)

print(f"FT: {F_t}")
print(f"FB: {F_b}")

alpha = 0.05

crit_f = round(st.f.ppf(1-alpha, df_t, df_e), 4)
print(f"Critical value: {crit_f}")
print()
if F_t > crit_f:
    print(f'There is sufficient evidence to conclude not all averages of music type and existence are similar')
else:
    print(f'There is NOT sufficient evidence to conclude not all averages of music type and existence are similar')
```

```
# poisson shit

val_setup = {
    'customers': [0, 1, 2, 3, 4, 5, 6],
    'freq': [1, 2, 9, 11, 14, 6, 9],
    'prob': [0.030196, 0.105691, .184959, .215785, .188812, .132169, .142386]
}

vals = pd.DataFrame(val_setup)
vals.set_index('customers', inplace=True)
vals
```

```
vals['expected'] = vals.prob * sum(vals.freq)
vals
```

```
# Ho: p_hati = p_hatj
# Ha: p_hati <> p_hatj

chi_sq_calc = sum( ( (vals.freq.values - vals.expected.values)**2) / vals.expected.values)
print(f"Test statistic: {round(chi_sq_calc, 4)}")

k = len(vals)
df_k = k - 1
alpha = 0.1

crit_val = round(st.chi2.ppf(1 - alpha, df_k), 4)
print(f"Critical value: {crit_val}")

claim = "that the proportions are not equally dispersed over the five possible responses"

if chi_sq_calc > crit_val:
    print(f"At alpha = {alpha}, there is sufficient evidence to state {claim}")
else:
    print(f"At alpha = {alpha}, there is not sufficient evidence to state {claim}")
```
