```
# 1) 
# I would like to select 3 paperback books from a list of 11 books to take on vacation. 
# How many different sets of three books can I choose?

select = 3
total = 11

total_combos = math.factorial(total) / (math.factorial(select) * math.factorial(total-select))
print(f"Total combtination sets of 3 available: {total_combos}")
```

```
# 2)
# Suppose there is a 90% chance that a test for performance-enhancing drugs will provide a positive test for an athlete who 
# actually took them.  In addition, there is a 15% chance that the same test will provide a positive test for an athlete who did 
# not take them.  Assume that 8% of the athlete population is currently taking performance-enhancing drugs.  
# Determine the probability that a randomly selected athlete who tests positive is actually a user of performance-enhancing drugs

p_user = 0.08
p_nuser = 1 - p_user
# positive result
p_user_given_pos = 0.9
# false positive, not catching the drug user
p_user_given_neg = 0.15

# bayes
p_user_given_pos = (p_user_given_pos * p_user) / ((p_user_given_pos * p_user) + (p_user_given_neg * p_nuser)) ## all results given user
    
print(f"P(User|Positive test) = {round(p_user_given_pos,3)}")
```

```
# 3)
# Costco offers two types of memberships; Gold Star members and Executive members.  Suppose that the probability of having a 
# check-out total of $200 or less is 0.58   Further suppose the probability of having a check-out total of $200 or less for 
# Gold Star members is 0.8148.  If the probability of being a Gold Star member or having check-out total of $200 or less is 0.68, 
# determine the probability of being a Gold Star member.

p_200_less = 0.58
p_200_less_given_gold = 0.8148
p_gold_or_checkout_min = 0.68

# algebra
# p_200_less_given_gold = p_gold_and_200_less / p_gold
# p_gold_and_200_less = p_200_less * p_gold
# p_gold_or_200_less = p_200_less + (1)p_gold - (p_200_less * p_gold) # 1 to note coefficient as we calculate p_gold

p_gold_and_200_less = p_gold_or_checkout_min - p_200_less
p_gold = p_gold_and_200_less / (1 - p_200_less_given_gold) 

print(f"P(Gold Star member) = {round(p_gold, 4)}")
```
