# Hypothesis testing

### Hypothesis tests

```
# i.
# Residents want to know about the speed of their 30 mph road. 
# Our hypotheses are:
#   Ho: mu = 30
#   Ha: mu > 30
```

```
# ii. 
alpha = 0.05
n = 15
# get df to use t tests since n < 30
df = n - 1

# get critical value
cv = round(st.t.ppf(alpha, df), 4)
print(f"Critical value: {cv}")
```

```
#  iv. 
# Ho : mu = 40
# Ha: mu > 40

n = 25
t_stat = -2.27
alpha = 0.1
df = n - 1

# get critical value, get absolute value since greater than
cv = abs(round(st.t.ppf(alpha, df), 4))

print(f"Critical value: {cv}")
```

```
# We can conclude there is not sufficient evidence to state the population mean can exceed 40 based on the sample mean.
```

```
p_val = round(st.t.cdf(t_stat, df), 4)
# do 1 - p_val since evaluating greater than
print(f"P-value: {1 - p_val}")
```

# Confidence interval

### Confidence intervals - all correct

```
# i.
n = 537
p = 0.53

# we use our large population proportion confidencce intervals for this problem

mean = n
var = round((p * (1 - p)) / n, 4)
avg = n * p

print(f"Mean: {mean}")
print(f"Var: {var}")
print(f"Average: {avg}")
```

```
# ii.

ci = 0.90
sample = 12
df = sample - 1

# we use the t table for this problem
alpha = (1 - ci) / 2

t_val = round(st.t.ppf(alpha, df), 4)
print(f"T value: {t_val}")
```
