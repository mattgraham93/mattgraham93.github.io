```r
# https://r-lang.com/mode-in-r/
# need to learn more about how this works and test with strings

getmode <- function(v) {
    uniqv <- unique(v)
    uniqv[which.max(tabulate(match(v, uniqv)))]
}

result <- getmode(values)
result
```
