6.3 (equal variances)

Conduct a test of H0: ml >= m2-2.3 versus Ha: ml < m2-2.3 for the sample data summarized here. Use a=.01 in reaching your conclusions

            
    Population  1       2
    n           13      21
    x_bar       50.3    58.6
    sd          7.23    6.98

```
# Ho: mu1 - mu2 >= -2.3
# Ha: mu1 - mu2 < -2.3

x_bar1 = 50.3
sd1 = 7.23
var1 = sd1**2
n1 = 13

x_bar2 = 58.6
sd2 = 6.98
var2 = sd2**2
n2 = 21

alpha = 0.01
test_diff = -2.3

# calculate degrees of freedom
df = n1 + n2 - 2
print(f"Degrees of freedom: {df}")

# get critical value
crit_val = round(st.t.ppf(alpha, df), 4)
print(f"Critical value: {crit_val}")

# calculate difference in mean
x_bar_diff = round(x_bar1 - x_bar2, 4)
print(f"Difference in means: {x_bar_diff}")

# get pooled sd
sd_pool = round(math.sqrt( ( ( (n1 - 1) * var1) + ( (n2 - 1) * var2)) / df), 4)
print(f"Pooled variance: {sd_pool}")

# setup numerator and denomenator
numer = round(x_bar_diff - test_diff)
denom = round(sd_pool * math.sqrt( (1/n1) + (1/n2) ), 4)

t_val = round(numer / denom, 4)
print(f"T-value: {t_val}")

p_val = round(st.t.cdf(t_val, df), 4)
print(f"P value: {p_val}")
```

At alpha=0.01, there is not sufficient evidence to conclude that the mean difference between population 1 and population 2 is greater than or equal to -2.3.

```
# a. In order for the discharge to have an impact on fish health, there needs to be MORE THAN an .5 ppm reduction in the 
# dissolved oxygen. Do the data provide sufficient evidence that there is a large enough reduction in the mean dissolved 
# oxygen between the upstream and downstream water in the river to impact the health of the fish? Use a=.01

#: Ho: x_bar1 - x_bar2 > 0.5
#: Ha: x_bar1 - x_bar2 < 0.5


x_bar1 = 4.9076
sd1 = 0.1870
var1 = sd1**2
n1 = 15

x_bar2 = 3.74
sd2 = 0.3203
var2 = sd2**2
n2 = 15

df_num = ((sd1**2 / n1) + (sd2**2 / n2)) ** 2
df_denom = ( ( (sd1**2 / n1) ** 2) / (n1 - 1) ) + ( ( (sd2**2 / n2) ** 2) / (n2 - 1) )

# always round down for degrees of freedom
df = math.floor(df_num / df_denom)
print(f"Degrees of freedom: {df}")

alpha = 0.01
test_diff = 0.5

# get critical value
crit_val = abs(round(st.t.ppf(alpha, df), 4))
print(f"Critical value: {crit_val}")

# calculate difference in mean
x_bar_diff = round(x_bar1 - x_bar2, 4)
print(f"Difference in means: {x_bar_diff}")

# setup numerator and denomenator
numer = round(x_bar_diff - test_diff, 4)
denom = round(math.sqrt( (var1/n1) + (var2/n2) ), 4)

t_val = round(numer / denom, 4)
print(f"T-value: {t_val}")

p_val = round(1 - st.t.cdf(t_val, df), 7)
print(f"P value: {p_val}")
```

A random sample of eight pairs of twins was randomly assigned to treatment A or
treatment B. The data are given in the following table.

    Twins           1       2       3       4       5       6       7       8

    Treatment B     43.5    43.8    53.7    43.9    54.4    54.7    45.2    34.4

```
# Is there significant evidence that the two treatments differ using an a=.05 paired t test. 

# Ho: mu1 - mu2 = 0
# Ha: mu1 - mu2 != 0

t_a = [48.3, 44.6, 49.7, 40.5, 54.3, 55.6, 45.8, 35.4]
t_b = [43.5, 43.8, 53.7, 43.9, 54.4, 54.7, 45.2, 34.4]

t_diff = []
i = 0
test_diff = 0

alpha = 0.05 / 2

while i < len(t_a):
    t_diff.append(t_a[i] - t_b[i])
    i += 1
# print(t_diff)

y_bar1 = round(np.average(t_a),4)
# print(y_bar1)
y_bar2 = round(np.average(t_b),4)
# print(y_bar2)
y_bar_diff = round(np.average(t_diff),4)
print(f"Difference average: {y_bar_diff}")

# calculated
# print(y_bar_diff)
# estimated
# print(y_bar1 - y_bar2)

n1 = len(t_a)
n2 = len(t_b)
n_diff = len(t_diff)

var_diff = round( (np.sum(np.square(t_diff)) - ( ( np.sum(t_diff)**2) / n_diff)) / (n_diff - 1), 4)
print(f"Difference variance: {var_diff}")

sd1 = round(float(np.std(t_a)), 4)
# print(sd1)
sd2 =round(float(np.std(t_b)), 4)
# print(sd2)
sd_diff = math.sqrt(var_diff)

df_diff = n_diff - 1

# get critical value
crit_val = round(st.t.ppf(alpha, df_diff), 4)
print(f"Critical value: {crit_val}")

# get test statistic
t_val = round((n_diff - test_diff) / (sd_diff / math.sqrt(n_diff)), 4)
print(f"T value: {t_val}")

# get p-value
p_val = st.t.cdf(t_val, df_diff)
print(f"P-value: {round(1 - p_val, 5)}")
# decide
```

We can conclude there at an alpha = 0.05 and two-tail test that there is sufficient evidence in the two treatment groups, A and B. We can confidently state that the difference between population A and population B exists, though not by how much.

Sludge is a dried product remaining from processed sewage and is often used as a fertilizer on agriculture crops. If the sludge contains a high concentration of certain heavy metals,
such as nickel, the nickel may reach such a concentration in the crops that it becomes a danger
to the consumer of the crops. A new method of processing sewage has been developed, and an
experiment is conducted to evaluate its effectiveness in removing heavy metals. Sewage of a
known concentration of nickel is treated using both the new and the old methods. One hundred
tomato plants were randomly assigned to pots containing sewage sludge processed by one of the
two methods. The tomatoes harvested from the plants were evaluated to determine if the nickel
was at a toxic level. The results are as follows:

    Level of Nickel
    Treatment   Toxic   Nontoxic    Total
    New         5       45          50
    Old         9       41          50
    Total       14      86          100

```
new_tox = 5
old_tox = 9

new_nontox = 45
old_nontox = 41

total_tox = new_tox + old_tox
total_nontox = new_nontox + old_nontox
total_new = new_tox + new_nontox
total_old = old_tox + old_nontox

total = total_nontox + total_tox

print(f"Totals: Tox - {total_tox}; Nontox - {total_nontox}; Total - {total}")

p_old_tox = p_hat1 = round(old_tox / total, 4)
p_new_tox = p_hat2 = round(new_tox / total_new, 4)
print(f"P old tox: {p_old_tox}; P new tox: {p_new_tox}")
```

```
# b. Is there significant evidence (a=.05) that the new treatment would produce a lower proportion of 
# plants having a toxic level of nickel compared to the old treatment?

# Ho: p_new = p_old
# Ha: p_new < p_old

p_diff = round(p_hat1 - p_hat2, 4)
print(f"P diff: {p_diff}")

p_c = (new_tox + old_tox) / (total_new + total_old)
print(f"P-hat pool = {p_c}")

# get critical value
alpha = 0.05 / 2
crit_val = abs(round(st.norm.ppf(alpha), 4))
print(f"Critical value: {crit_val}")

test_diff = 0

# get test statistic
z_val = round((p_diff - test_diff) / math.sqrt(p_c* (1 - p_c) * ((1/total_new) + 1/total_old)), 4)
print(f"Z value: {z_val}")

# get p-value
p_val = round(st.norm.cdf(z_val), 4)
print(f"P value: {p_val}")

# decide
```

At an alpha = 0.05, we can conclude there is not sufficient evidence to state there is a difference between old and new sludge treatments.
