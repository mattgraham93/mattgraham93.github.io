```
import math
import scipy.stats as st
```

```
# 1
n = 36
x_bar = 21.32
s = 4.30

ci = 0.95
alpha = 1 - ci

two_tail = True

if two_tail:
    ci += alpha / 2

z_score = round(st.norm.ppf(ci), 4)
# print(f"Z score: {z_score}")

lower = round(x_bar - (z_score * (s/math.sqrt(n))), 4)
upper = round(x_bar + (z_score * (s/math.sqrt(n))), 4)

print(f"{(ci - alpha/2) * 100}% confidence interval: ({lower}, {upper})")
```

```
# The bureau's determination of an average wage of $22.57 makes sense with our confidence interval, but it is at
# the very high end. 
```

```
# 2
total_seats = 3500  # total capacity of basketball facility
sample = 400  # basketball fans
x = 260  # fans who purchased concessions

p_hat = x / sample

sd = (p_hat * (1-p_hat)) / sample

ci = 0.99
alpha = 1 - ci
two_tail = True

if two_tail:
    ci += alpha / 2

z_score = round(st.norm.ppf(ci), 4)
# print(f"Z score: {z_score}")

lower = round(p_hat - (z_score * (math.sqrt(sd))), 4)
upper = round(p_hat + (z_score * (math.sqrt(sd))), 4)

print(f"{(ci - alpha/2) * 100}% confidence interval: ({lower}, {upper})")
```

```
# We are 99% confident that the proportion of basketball fans who attend a game that purchase concessions is between 0.5886 and 0.7114
```

```
# 3

# Ho: mu = 73.3
# Ha: mu <> 73.3    "evidence to indicate number of days HAS CHANGED" --> not equal to

mu = 73.3 # days in advance travelers book airline reservations
n = 40 # travelers
x_bar = 71.4
sd = 5.3
alpha = 0.05
two_tail = True

if two_tail:
    alpha = alpha / 2

cv = round(st.norm.ppf(alpha), 4)
print(f"Critical value: {cv}")

test_stat = round((x_bar - mu) / (sd / math.sqrt(n)), 4)
print(f"Test statistic: {test_stat}")

p_val = round(st.norm.cdf(test_stat), 4)
print(f"P-val: {2 * p_val}") # ----> two tail means we do 2 * p_val
```

```
# There is sufficient evidence to conclude that the days before travelers booked travel has changed from 73.3 days recorded in 2008
```

3 - Marketing agency wants to know acceptance rate of chocolate flavored medicine. Calculate confidence interval

```
n = 120
x = 51
ci = 0.90
alpha = 1 - ci
p_hat = x / n

two_tail = True

if two_tail:
    ci += (alpha / 2)

z = round(st.norm.ppf(ci), 4)
print(z)

prod = math.sqrt((p_hat * (1 - p_hat)) / n)

# calculate confidence interval
lower = round(p_hat - (z * prod), 4)
upper = round(p_hat + (z * prod), 4)

print(f"90% confidence interval of proportion: ({lower},{upper})")
```

# Power

power = beta = type 2 error

power = beta = p(z < za) = p ( z < za + (muo - mua) / (sigma / math.sqrt(n)) ) <-- usually a negative number

```
# Ha: mu > 200
# what is power when mua = 210 is true?

muo = 200
n = 64
sigma = 100
mua = 210
alpha = 0.05

za = round(abs(st.norm.ppf(alpha)), 4)
# print(za)
tii_err = za + ( (muo - mua) / (sigma / math.sqrt(n)) )
print(f"P(Type II error) = {tii_err}")

power = round(1 - tii_err, 4)
print(f"Power = {power}")
```

```
# Ha: mu > 30
alpha = 0.05
s = 10
n = 36
mua = 34
muo = 30

# absolute value since greater than
za = abs(round(st.norm.ppf(alpha), 4))
# print(za)
z_tii_err = round(za + ( (muo - mua) / (s / math.sqrt(n)) ), 4)
tii_err = round(st.norm.cdf(z_tii_err), 4)
print(f"P(Type II error) = {tii_err}")

power = round(1 - tii_err, 4)
print(f"Power = {power}")
```
