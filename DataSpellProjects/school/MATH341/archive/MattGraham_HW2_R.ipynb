#### 4.45 - Binomial
In an attempt to decrease drunk driving, police set up vehicle checkpoints during the July 4 evening. The police randomly select vehicles to be stopped for “informational” checks. On 
a particular roadway, assume that 20% of all drivers have a blood alcohol level above the legal limit. For a random sample of 15 vehicles, compute the following probabilities

P(above legal limit) = 0.2, n=15

```r
# a. All 15 drivers will have a blood alcohol limit exceeding legal limit
val1 <- dbinom(15, 15, .2)
print(val1)
```

```r
# b. Exactly 6 people will exceed legal limit
val2 <- dbinom(6, 15, .2)
print(val2)
```

```r
# c. Of the 15, 6 or more will exceed the legal limit
val3 = 1 - pbinom(5, 15, .2)
print(val3)
```

```r
# visual for help
success <- 0:15
plot(success, dbinom(success, size=15, prob=0.2),type='h')
```

```r
# d. all drivers will be within legal limit
val4 <- dbinom(0, 15, .2)
print(val4)
```

#### 4.50 - Poisson

Customers arrive at a grocery store checkout at a rate of six per 30 minutes during the hours of 5 p.m. and 7 p.m. during the workweek. Let C be the number of customers arriving at the checkout during any 30-minute period of time. The management of the store wants to determine the frequency of the following events. Compute the probabilities of these events

```r
mins <- 30
customers <- 6

c = customers / mins
```

```r
plot(customers, )
```

```r
# a. No customers arrive
poi1 <- dpois(0, customers, c)
print(poi1)
```

```r
# b. More than 6 customers arrive
poi2 <- ppois(6, customers, c)
print(poi2)
```

```r
# c. At most three customers arrive
poi3 <- 1 - ppois(3, customers, c)
print(poi3)
```

#### Word 1 - Hypergeometric

A company needed to downsize one department having 30 people – 12 women and 18 men.  Ten people were laid off, and upper management said the layoffs were done randomly.  However of the 10 people laid off, 8 were women.  Using the hypergeomentric distribution find the probability of 8 or more women being laid off by chance alone.

```r
m <- 18
w <- 12
# total people
t <- m+w

# total laid off
lo <- 10
p_lo = lo / t

# use when graphing hypergeometric dist.
x <- seq(0:lo)

# set laid off proportions
p_fglo <- 0.8
p_mglo <- 0.2

# get total laid off folks
w_lo <- lo*p_fglo
m_lo <- lo*p_mglo

# Probability exactly 8 women are laid off by chance alone, need to make it great than
hyp1 <- dhyper(w_lo,t,lo,w_lo)
print(hyp1)
```

```r
# graph hypergeometric
y_dhyper <- dhyper(x,t,lo,w_lo)
plot(y_dhyper)
```
