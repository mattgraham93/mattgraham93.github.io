```python
# The standard start of our notebooks
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
import re
```

Load the data set. We are not going to need the labels ($iris.target$ is irrelevant; working only with $iris.data$), and we are not breaking the data into training and test sets.

```python
# Load the data, which is included in sklearn.
iris = load_iris()
print ('Iris feature names:', iris.feature_names)
X, Names = iris.data, iris.feature_names

##  Remove the " (cm)" part
pattern = r' \(cm\)'
for nn in np.arange(0,len(Names)):
    Names[nn] = re.sub (pattern, lambda x: "", Names[nn])
print (f"Cleaned up names: {Names}")

print (f"df shape: {X.shape}")

# Shuffle the data, but make sure that the features and accompanying lalen(bels stay in sync.
np.random.seed(0)  #  To ensure repeatability of results
shuffle = np.random.permutation(np.arange(X.shape[0]))
X = X[shuffle]
# X
```

# Your Task

Let's explore the first feature Sepal Length List.
1. Calculate the mean length using sum of all the values of the list / total number of all the values
2. Check if the mean is less than 5
3. Define a function to check the average value of the total. CANNOT use numpy.mean().
4. Write an if-else statement to check if the length is less than 5. If so return "Short", if not, return "Long".
5. Write a for-loop statement to display all the Short values. Tips: check the length of "Sepal Length" at first and then use the length to define the i in for loop.

```python
# get by column index
sep_len = X[:,0]
# sep_len
```

```python
# 1
mean_len = sep_len.sum() / len(sep_len)
print(f"Mean sepal length: {round(mean_len, 3)}cm")
```

```python
# 2
if mean_len < 5:
    print("Mean sepal length is less than 5")
elif mean_len > 5:
    print("Mean sepal length is greater than 5")
else:
    print("Mean = 5")
```

```python
# 3
def calc_mean(vals):
    # print("Used function")
    mean_len_new = vals.sum() / len(vals)
    return mean_len_new

print(f"Mean sepal length: {calc_mean(sep_len)}")
```

```python
# 4
if mean_len < 5:
    print("Short")
elif mean_len > 5:
    print("Long")
else:
    print("Average")
```

```python
# 5
i = 0
vals = []
# iterate through all sepal lengths
for leng in sep_len:
    if leng < 5:
        # append value to list
        vals.append(leng)

print(f"Values below 5: {vals}")
```
