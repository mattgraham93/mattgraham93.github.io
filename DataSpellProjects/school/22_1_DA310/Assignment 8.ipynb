```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```
# create dataframe from csv
df = pd.read_csv('HeartDisease.csv')
# print shape
print(f"Shape: {df.shape}")

# create columns object
col_obj = df.columns  

cols = []  
# iterate through col_obj to make list for graph titles
for col in col_obj:
    cols.append(col)

# show columns
print(f"Columns: {cols}")

# put to array for np operations
df_np = df.to_numpy()
# print(df_np.shape)
```

```
fig, axes = plt.subplots(nrows=1, ncols=df_np.shape[1], sharex=False, sharey=False, figsize = (16, 3))

for ii in np.arange(0, df_np.shape[1]):
    # print(ii)
    ax = axes[ii]
    # print(ax)
    ax.hist(df_np[:,ii])
    ax.set_title(cols[ii])

plt.show()
```

```
## Bad programming style (spaghetti code), but didn't want to add confusion about looping the features back 
# to the first one
fig, axes = plt.subplots(nrows=df_np.shape[1], ncols=df_np.shape[1], sharex=False, sharey=False, figsize = (16, 16))

for f1 in np.arange (df.shape[1]):
    for f2 in np.arange (df.shape[1]):
        if not f1 == f2:
            ax = axes[f1, f2]
            ax.scatter (df_np[:,f1], df_np[:,f2])
            ax.set_title (cols[f1] + " vs " + cols[f2])
            
plt.show()
```
