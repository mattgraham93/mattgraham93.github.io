```python
def get_mongo_creds():
    import json
    creds_loc = r"C:\auth\mongo_auth.json"

    with open(creds_loc) as f:
        data = json.load(f)
        conn_str = data['conn_string']
        username = data['username']
        password = data['password']

    return conn_str, username, password
```

```python
def get_mongodb():
    from pymongo import MongoClient
    import pymongo

    conn_str, username, password = get_mongo_creds()
    
    client = pymongo.MongoClient(conn_str)
    return client.test
```
