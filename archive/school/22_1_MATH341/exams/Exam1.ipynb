```
import math
import scipy.stats as st
```

```
n = 23
mu = 60
x_bar = 54.95
s = 16.87
df = n - 1
alpha = 0.05
two_tail = True

if two_tail:
    alpha = alpha / 2
# calculate critical value
cv = round(st.t.ppf(alpha, df), 4)
print(f"Critical value: {cv}")

# calculate test statistic
t_score = round((x_bar - mu) / (s / math.sqrt(n)), 4)
print(f"Z score: {t_score}")

p_val = round(st.t.cdf(t_score, df), 4)
print(f"P-value: {p_val*2}")
```

```
n = 350
x = 10

p = 0.05
p_hat = round(x / n, 4)
print(f"P_hat: {p_hat}")
alpha = 0.10

# determine test type
np = n * p #  needs to be > 5 for z
nq = n * (1 - p) #  AND needs to be > 5 for z

# print(f"{np}, {nq}")
# since np and nq are both greater than 5, we can use t-tests 
s = math.sqrt( p * (1 - p) / n)

# find critical value
cv = round(st.norm.ppf(alpha), 4)
print(f"Critical value: {cv}")

# calculate test statistic
z_score =  round((p_hat - p) / s, 4)
print(f"Z score: {z_score}")

# calculate p-value
p_val = round(st.norm.cdf(z_score), 4)
print(f"P-value: {p_val}")
```

```
n = 100
x_bar = 12.09
s = 0.2

ci = 0.95
alpha = 1 - ci

two_tail = True

if two_tail:
    ci += alpha / 2

z_score = round(st.norm.ppf(ci), 4)
print(f"Z score: {z_score}")

lower = round(x_bar - (z_score * (s/math.sqrt(n))), 4)
upper = round(x_bar + (z_score * (s/math.sqrt(n))), 4)

print(f"{(ci - alpha/2) * 100}% confidence interval: ({lower}, {upper})")
```

```
mu = 130 # days in advance travelers book airline reservations
n = 44 # travelers
x_bar = 136.98
sd = 15.43
alpha = 0.01
two_tail = False

if two_tail:
    alpha = alpha / 2

cv = round(st.norm.ppf(alpha), 4)
print(f"Critical value: {cv}")

test_stat = round((x_bar - mu) / (sd / math.sqrt(n)), 4)
print(f"Test statistic: {test_stat}")

p_val = round(st.norm.cdf(test_stat), 4)
print(f"P-val: {p_val}") # ----> two tail means we do 2 * p_val
```

```
# a - 99% CI
n = 20
df = n - 1
x_bar = 126.2
s = math.sqrt(83.51)
ci = 0.99
alpha = 1 - ci

# define product to multiply for t-confidence intervals
prod = math.sqrt(s**2 / n)

two_tail = True

if two_tail:
    alpha = alpha / 2
    ci += alpha

t = round(st.t.ppf(ci, df), 4)
print(t)

lower = x_bar - (t * prod)
upper = x_bar + (t * prod)

print(f"99% confidence interval: ({round(lower, 4)} lbs, {round(upper, 4)} lbs)")
```

```
#  Ha: mu < 130
# what is power when mua = 128 is true?

muo = 130
n = 44
sigma = 15.43
mua = 128
alpha = 0.01

za = round(st.norm.ppf(alpha), 4)
print(za)
z_tii_err = za + ( (muo - mua) / (sigma / math.sqrt(n)) )
tii_err = round(st.norm.cdf(z_tii_err), 4)
print(f"P(Type II error) = {tii_err}")


power = round(1 - tii_err, 4)
print(f"Power = {power}")
```
