```
import math
import scipy.stats as st
import numpy as np
```

```
# 1
# this is a parid difference becase we are using 2 sets of identical cars to test mpg differences in fuel additive

# Ho: mud = 2 mpg
# Ha: mud < 2 mpg

test_diff = 2

n1 = n2 = n_diff = 25  # 1 = with additive 2 = w/o add
sd1 = 3.26
mean1 = 18.16

sd2 = 3.10
mean2 = 16.27


mean_diff = round(mean1 - mean2, 4)
sd_diff = 0.31  # cannot be calculated from the 2 sd's from group. will be given. if not given this, it is independent
print(f"Mean diff: {mean_diff}")

t = round((mean_diff - test_diff) / (sd_diff / math.sqrt(n_diff)), 4)
print(f"t: {t}")


df_diff = n_diff - 1
alpha = 0.10

crit_val = round(st.t.ppf(alpha, df_diff), 4)
print(f"Critical value: {crit_val}\n")

claim = "the average increase the increase is <2 mpg"

################################### UPDATE HERE TO DETERMINE CASE 
if(t < crit_val):
    print(f"At alpha={alpha}, we can conclude there is sufficient \nevidence to refute {claim}")
else:
    print(f"At alpha={alpha}, we can conclude there is not sufficient evidence to {claim}")
```

```
# 2
# small independent sample averages, equal variance

n1 = 16
var1 = 15.65
mean1 = 14.15
df1 = n1 - 1

n2 = 16
var2 = 25.92
mean2 = 14.61
df2 = n2 - 1

mean_diff = mean1 - mean2
var_diff = 16.02  # given

# construct 95% CI

alpha = 0.05 / 2

# check variance > F- test
print(f"---- Testing for equal variance ----")
ratio = round(var2 / var1, 4)  # LARGE VALUE GOES ON TOP
print(f"Ratio: {ratio}")

crit_val = round(st.f.ppf(1 - alpha, df1, df2), 4)
print(f"F Critical value: {crit_val}")

f_val = round(var1 / var2, 4)
print(f"F Test statistic: {f_val}\n")

p_val = round(1 - st.f.cdf(f_val,df1,df2), 4)
print(f"P value: {p_val}")

if f_val > crit_val:
    print(f"Assume unequal variance")
    df_num = ((sd1**2 / n1) + (sd2**2 / n2)) ** 2
    df_denom = ( ( (sd1**2 / n1) ** 2) / (n1 - 1) ) + ( ( (sd2**2 / n2) ** 2) / (n2 - 1) )
    # always round down for degrees of freedom
    df = math.floor(df_num / df_denom)
    print(f"Degrees of freedom: {df}")
else:
    print(f"Assume equal variance")
    df = n1 + n2 - 2

# insufficient evidence to conclude unequal variances -> equal variance
print("-------------------------")
print(f"Degrees of freedom: {df}")

crit_val = round(abs(st.t.ppf(alpha, df)), 4)
# print(f"Critical value: {crit_val}")

s_pool = round(math.sqrt( ( ( (n1 - 1) * var1) + ( (n2 - 1) * var2)) / df), 4)
print(f"Pooled variance: {s_pool}")

y_diff = round(mean1 - mean2, 4)
# print(f"Mean diff: {y_diff}")

prod = round(crit_val * s_pool * math.sqrt( (1/n1) + (1/n2) ), 4)
print(f"Product: {prod}\n")

lower = round(y_diff - prod, 4)
upper = round(y_diff + prod, 4)

print(f"{round((1 - 2 *alpha) * 100, 0)}% confidence interval: ({lower}, {upper})")

# 2.832 higher for A and 3.752 higher for B
```

```
# 3
# PAIRED DIFFERENCE > same person since we are comparing results of same person over period of time

# 90% confidence interval on mu_diff

alpha = 0.10 / 2

n1 = n2 = n_diff = 8

mean1 = 69.12
sd1 = 8.49
df1 = n1 - 1

sd1 = 8.98
mean2 = 79
df2 = n2 - 1

mean_diff = mean1 - mean2
sd_diff = 1.55  # given
var_diff = sd_diff**2

df_diff = n1 - 1

crit_val = round(abs(st.t.ppf(alpha, df_diff)), 4)
print(f"Critical value: {crit_val}")


prod = round(math.sqrt( var_diff / n_diff ), 4)
print(f"Multiplier: {round(prod * crit_val, 4)}")

lower = round(mean_diff - (crit_val * prod), 4)
upper = round(mean_diff + (crit_val * prod), 4)

print(f"{round((1 - (alpha * 2)) * 100, 1)}% confidence interval: ({lower}, {upper})")
```

```
# looking at proportion of students who live more than 20 miles away from campus

# is there sufficient evidence to conclude prop of males who live more than 20 miles away
# is higher than females who live more than 20 miles away

alpha = 0.01

# Ho: mu_male - mu_female = 0
# Ha: mu_male - mu_female > 0

p_male = 22 / 100
p_female = 15 / 100

p_c = round((22 + 15) / 200, 4)

p_diff = round(p_male - p_female, 5)
print(f"P diff: {p_diff}")

crit_val = abs(round(st.norm.ppf(alpha), 4))
print(f"Critical value: {crit_val}")

test_diff = 0

# denom = math.sqrt(p_c* (1 - p_c) * ((1/total_new) + 1/total_old))
# print(denom)

# get test statistic
z_val = round((p_diff - test_diff) / math.sqrt(p_c * (1 - p_c) * ((1/100) + 1/100)), 4)
print(f"Z value: {z_val}")

# get p-value
p_val = round(st.norm.cdf(z_val), 4)
print(f"P value: {p_val}")

# insufficient evidence to conclude males living more than 20 mi away exceeds females living 20 mi away
claim = "there is no difference in proportion among male and female students who live more than 20 miles away from campus"
################################### UPDATE HERE TO DETERMINE CASE 
if z_val > crit_val:
    print(f"At alpha={alpha}, we can conclude there is sufficient evidence to refute \n{claim}")
else:
    print(f"At alpha={alpha}, we can conclude there is not sufficient evidence refute \n{claim}")
```
