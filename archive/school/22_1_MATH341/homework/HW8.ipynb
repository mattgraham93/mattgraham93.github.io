```
# setup table
vals = pd.DataFrame(columns=['Item', 'Control','Piano','Computer','Instructor'])

loc1_mean = 7.00
loc1_var = 7.341
vals = vals.append({'Item': 'Mean', 'Control': -0.112, 'Piano': 3.792, 'Computer':4.764, 'Instructor':7.4}, ignore_index=True)
vals = vals.append({'Item': 'SD', 'Control': 3.3091, 'Piano': 2.7463, 'Computer':3.1399, 'Instructor':3.3656}, ignore_index=True)
vals = vals.append({'Item': 'Size', 'Control': 25, 'Piano': 25, 'Computer':25, 'Instructor':25}, ignore_index=True)
vals.set_index('Item', inplace=True)
vals_t = vals.T
vals_t
```

```
# a. Conduct an analysis of variance, and summarize your results in an AOV table

t = int(vals_t.Size.unique())  # this will also be len(df) if unique rows
print(f"t: {t}")

N = vals_t.shape[0] * t
print(f"N: {N}")

b = len(vals_t) # total number of buckets
print(f"b: {b}\n")

df_b = b - 1
df_t = t - 1
df_e = (t - 1) * (b - 1)
df_N = N - 1

print(f"DFB: {df_b}\nDFT: {df_t}\nDFE: {df_e}\nDFN: {df_N}\n")

x_bar = round(sum(vals_t.Mean.values * vals_t.Size.values) / sum(vals_t.Size.values), 4)
print(f"x_bar = {x_bar}")

sst = round(b * sum((vals_t.Mean - x_bar)**2), 4)
print(f"SST: {sst}")

ssb = round(sum(vals_t.Size * (vals_t.Mean - x_bar)**2), 4)
print(f"SSB: {ssb}")

ssw = round(sum((vals_t.Size - 1) * vals_t.SD**2), 4)
print(f"SSW: {ssw}")

msb = round(ssb / (t - 1), 4)
print(f"MSB: {msb}")

msw = round(ssw / (N - t), 4)
print(f"MSW: {msw}")

f = round(msb / msw, 4)
print(f"F-value: {f}")
```

```
# b. Test the research hypothesis that there is a difference in the effectiveness means of the methods of instruction. Use a=.05
alpha = 0.05

# Critical value
q_val = crit_val = round(st.f.ppf(1 - alpha, df_b, df_t),4)
print(f"Critical value: {crit_val}")

# Get p-value
p_val = round(1 - st.f.cdf(f, df_b, df_t), 4)
print(f"P value: {p_val}")

# At alpha = 0.05, there is not sufficient evidence to conclude there is a difference in effectiveness across the different methods
# of teaching music. 
```

```
# c. Apply a multiple-comparison procedure to determine pairwise differences in the three instructional methods. Use a=.05
alpha = 0.05

diffs = []
diffs.append(round(abs(vals_t.loc["Control", "Mean"] - vals_t.loc["Piano", "Mean"]), 4))
diffs.append(round(abs(vals_t.loc["Control", "Mean"] - vals_t.loc["Computer", "Mean"]), 4))
diffs.append(round(abs(vals_t.loc["Control", "Mean"] - vals_t.loc["Instructor", "Mean"]), 4))

print(f"Differences: {diffs}")

ranges = []
ranges.append(round(q_val * math.sqrt((msw / 2) * ((1 / vals_t.loc["Control", "Size"]) + (1 / vals_t.loc["Piano", "Size"]))), 4))
ranges.append(round(q_val * math.sqrt((msw / 2) * ((1 / vals_t.loc["Control", "Size"]) + (1 / vals_t.loc["Computer", "Size"]))), 4))
ranges.append(round(q_val * math.sqrt((msw / 2) * ((1 / vals_t.loc["Control", "Size"]) + (1 / vals_t.loc["Instructor", "Size"]))), 4))

print(f"Critical ranges: {ranges}")
```

```
# setup table
cols = ['Type/Genre', 'H1', 'H2', 'H3', 'H4', 'H5', 'H6', 'H7']
music = pd.DataFrame(columns=cols)

music = music.append({'Type/Genre':'N/A', 'H1':20, 'H2':17, 'H3':24, 'H4':20, 'H5':22, 'H6':25, 'H7':18}, ignore_index=True)
music = music.append({'Type/Genre':'Hard', 'H1':20, 'H2':18, 'H3':23, 'H4':18, 'H5':21, 'H6':22, 'H7':19}, ignore_index=True)
music = music.append({'Type/Genre':'Classical', 'H1':24, 'H2':20, 'H3':27, 'H4':22, 'H5':24, 'H6':28, 'H7':17}, ignore_index=True)
music.set_index('Type/Genre', inplace=True)
music
```

```
x_bar = round(sum(music.mean(axis=1)) / len(music), 4)
print(f"x_bar = {x_bar}")

# ANOVA table items needed
N = music.shape[0] * music.shape[1]
print(f"N: {N}\n")

tss = round(sum(sum(music.values**2)) - (music.values.sum()**2 / N), 4)
print(f"TSS: {tss}")

b = len(music.columns)
sst = round(b * sum((music.mean(axis=1) - x_bar)**2), 4)
print(f"SST: {sst}")

t = len(music)
ssb = round(t * sum((music.mean(axis=0) - x_bar)**2), 4)
print(f"SSB: {ssb}")

sse = round(tss - sst - ssb, 4)
print(f"SSE: {sse}\n")

df_b = b - 1
df_t = t - 1
df_e = (t - 1) * (b - 1)
df_N = N - 1

print(f"DFB: {df_b}\nDFT: {df_t}\nDFE: {df_e}\nDFN: {df_N}\n")

mst = round(sst / df_t, 4)
print(f"MST: {mst}")

msb = round(ssb / df_b, 4)
print(f"MSB: {msb}")

mse = round(sse / df_e, 4)
print(f"MSE: {mse}\n")

F_t = round(mst / mse, 4)
F_b = round(msb / mse, 4)

print(f"FT: {F_t}")
print(f"FB: {F_b}")
```

```
# b. Are there differences in the mean typing efficiencies for the three types of music? Use a=.05

# get crit val
alpha = 0.05
crit_val = round(st.f.ppf(1-alpha, df_t, df_e), 4)
print(f"Critical value: {crit_val}")

# Get p-value
p_val = round(1 - st.f.cdf(F_t, df_t, df_e), 4)
print(f"P-value: {p_val}")

# At alpha = 0.05, we can conclude that there is enough evidence to conclude that there are typing efficiencies for the 
# three different types of music. 
```
