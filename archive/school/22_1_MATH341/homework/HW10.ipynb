 Using data from 11.12 obtain the 95% prediction interval when pectin = 2.5% 					
						
	
	  	  	  	  	  	 
						
	11.12 					
		x 	y 	x2      xy      y2
	S 	18 	750.87  42      1237.47 47971.127

```python
# Set variables for interval setting

sm_x = 18
sm_xx = 42
sm_y = 750.87
sm_yy = 47971.127
sm_xy = 1237.47

n = 12
y_bar = sm_y / n
x_bar = sm_x / n


b1 = round((sm_xy - ((sm_x * sm_y) / n)) / (sm_xx - (sm_x**2 / n)), 4)
print(f"coefficient: {b1}")

b0 = round(y_bar - (b1 * x_bar), 4)
print(f"intercept: {b0}")

print(f"Linear equation: prediction = {b0} + ({b1} * (variable))")


s_yx2 = round((sm_yy  - ((sm_xy**2) / sm_xx)) / (n - 2), 4)
print(f"s_yx2: {s_yx2}")
```

```python
x_pred = 2.5
pred = b0 + (b1 * x_pred)
print(f"Prediction: {pred}")
```

```python
alpha = 0.05 / 2
deg_free = n - 2

crit = st.t.ppf(1 - alpha, deg_free)

pref = b0 + (b1 * x_pred)
prod = crit * math.sqrt( s_yx2 * (1 + (1/n) + (((x_pred - x_bar)**2)/sm_xx)))

lower = pref - prod
upper = pref + prod

print(f"{round((1-2*alpha) * 100, 0)}% confidence interval: ({round(lower, 4)}, {round(upper, 4)})")
```
