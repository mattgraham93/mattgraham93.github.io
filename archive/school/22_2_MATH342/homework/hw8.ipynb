```python
df_1 = 9
df_2 = 40
alpha = .05

crit_val = st.f.ppf(1 - alpha, df_1, df_2)
print(f"Critical value: {crit_val}")
```

```python
ms_batch = 5.738
ms_error = 4.558

batches = i = 10
selection = j = 5
n = i * j

ms_tau = (ms_batch - ms_error) / j
print(f"ms_tau: {ms_tau}")

ms_y = ms_error + ms_tau
print(f"ms_y: {ms_y}")

prop_var = ms_tau / ms_y
print(f"Proportion of variability: {prop_var}")
```

### 17.2

```python
alpha = 0.05 / 2
y_base_avg = 5.242

t_val = st.t.ppf(1 - alpha, df_1)
print(f"t-value: {t_val}")

prod = t_val * math.sqrt(ms_batch / n)

lower = y_base_avg - prod
upper = y_base_avg + prod

print(f"95% confidence interval: ({lower}, {upper})")
```

```python
ms_subj = 10.4952
ms_anlyst = 0.1665
ms_error = 0.0195

subjects = i = 10
anlysts = j = 5
n = i * j


# Checking for subjects
ms_tau_subj = (ms_subj - ms_error) / j
print(f"ms_tau (Subject): {ms_tau_subj}")

ms_y_subj = ms_error + ms_tau_subj
print(f"ms_y (Subject): {ms_y_subj}")

prop_var_subj = ms_tau_subj / ms_y_subj
print(f"Proportion of variability (Subject): {prop_var_subj}")


# Checking for analysts
ms_tau_anst = (ms_anlyst - ms_error) / i
print(f"ms_tau (Analyst): {ms_tau_anst}")

ms_y_anst = ms_error + ms_tau_anst
print(f"ms_y: {ms_y_anst}")

prop_var_ants = ms_tau_anst / ms_y_subj
print(f"Proportion of variability (Analyst): {prop_var_ants}")


# Checking for error
# ms_tau_subj = (ms_subj - ms_error) / j
print(f"ms_tau (error): {ms_error}")

prop_var_subj = 1 - (prop_var_subj + prop_var_ants)
print(f"Proportion of variability (Subject): {prop_var_subj}")
```

```python
alpha = 0.05 / 2
df_1 = 9
y_base_avg = 10.25

n = subjects * anlysts
print(f"N: {n}")

t_val = st.t.ppf(1 - alpha, df_1)
print(f"t-value: {t_val}")

prod = t_val * math.sqrt((ms_anlyst + ms_subj - ms_error) / n)
print(f"Product: {prod}")

lower = y_base_avg - prod
upper = y_base_avg + prod

print(f"95% confidence interval: ({lower}, {upper})")
```
