# counting and distributions

#### imports

```
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

### counting
combinations and permutations

```
st = "abcde"
per = itertools.permutations(st)

for val in per:
    print(*val)
```

```
com_wr = itertools.combinations_with_replacement(st,4)
for val in com_wr:
    print(*val)
```

```
com = itertools.combinations(st,2)
for val in com:
    print(*val)
```

```
com = itertools.combinations(range(5),3)

count = 0
for val in com:
    count += 1
print(count)
```

### Poisson

```
fig, ax = plt.subplots(1,1)

mu = 0.6
mean, var, skew, kurt = stats.poisson.stats(mu, moments='mvsk')

x = np.arange(stats.poisson.ppf(0.01, mu),
              stats.poisson.ppf(0.99, mu))
ax.plot(x, stats.poisson.pmf(x, mu), 'bo', ms=8, label='poisson pmf')
ax. vlines(x, 0, stats.poisson.pmf(x, mu), colors='b', lw=5, alpha=0.5)

rv = stats.poisson(mu)

ax.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1, label='frozen pmf')
ax.legend(loc='best', frameon=False)
plt.show()
```

### hypergeometric

```

# hypergeometric setup
[M, n, N] = [20, 7, 12]

rv = stats.hypergeom(M, n, N)
x = np.arange(0, n+1)
pmf_dogs = rv.pmf(x)
```

```
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, pmf_dogs, 'bo')
ax.vlines(x, 0, pmf_dogs, lw=2)
ax.set_xlabel('# of dogs in our group of chosen animals')
ax.set_ylabel('hypergeom PMF')
plt.show()
```

```
low = 2
high = 4
prob = 0

while low <= high:
    prob += rv.pmf(low)
    low += 1

prob
```

```
cdf_test = rv.cdf(4) - rv.cdf(1)
cdf_test
```
