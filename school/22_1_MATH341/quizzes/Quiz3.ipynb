## #1

```
# 1: Airline wants to lower weight of check-in from 50lbs. Estimate weight per bag of passengers

n = 23
df = n - 1
x_bar = 38.5
s = 3.79
```

```
# a - 95% CI
ci = 0.95
alpha = 1 - ci

# define product to multiply for t-confidence intervals
prod = math.sqrt(s**2 / n)

two_tail = True

if two_tail:
    alpha = alpha / 2
    ci += alpha

t = round(st.t.ppf(ci, df), 4)
# print(t)

lower = x_bar - (t * prod)
upper = x_bar + (t * prod)

print(f"95% confidence interval: ({round(lower, 3)} lbs, {round(upper, 3)} lbs)")
```

```
# We can state that the airlines are safely able to lower their weight check-in limit. We are 95% confident that the average
# flier will check-in a bag between 36.861 lbs and 40.139 lbs.
```

## #2

```
# 2: Marketing company scared of cell phone exclusive users. Conducted mail survey. 8% of 250 respondents say they
#     exclusively used cell phones. Calculate 90% CI

p = 0.08
n = 250

ci = 0.9
alpha = 1 - ci
prod = math.sqrt( (p * (1 - p)) / n)

two_tail = True

if two_tail:
    ci += alpha / 2

z = st.norm.ppf(ci)

lower = p - (z * prod)
upper = p + (z * prod)

print(f"90% confidence interval: ({round(lower* 100, 4)}%, {round(upper * 100, 4)}%)")
```

```
# We can conclude that we are 90% confident that the population of people who would respond to a mailed-in survey,
# targeting cell-phone exclusive users, is between 5.1777% and 10.8223%. I would hypothesize that this is a very bad
# way to look for these users. I would recommend social media compaigns. 
```

```
# 3: Company wants to lower health care coverage for employees (nice). Claim: Avg amount of insurance claims filed
#   by employees < $250. Company sampled 100 people. Test stat = 2.31

# REFUTE THE CLAIM MEANS THE OPPOSITE OF INITIAL CLAIM

# a. hypothesis test
alpha = 0.05
t_stat = 2.31

# get critical value
cv = round(st.norm.ppf(alpha), 4)
print(f"Critical value: {cv}")
```

```
# Since our test staistic is 2.31, and we can only accept the claim when the value is LOWER than -1.6449 - we do not have
# sufficient evidence to conclude that the average health care claims among employees is lower than 250. In fact, we can
# state that there is sufficient evidence to say that it is higher than $250.
```

```
p_val = st.norm.cdf(t_stat)
print(f"P-value: {round(1 - p_val, 4)}")
```
