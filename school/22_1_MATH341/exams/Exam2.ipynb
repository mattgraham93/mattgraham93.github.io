```
import math
import scipy.stats as st
import numpy as np
```

```
# 1
y_bar1 = 141.3
var1 = 46.11
sd1 = math.sqrt(var1)
n1 = 32

y_bar2 = 127.4
var2 = 31.92
sd2 = math.sqrt(var2)
n2 = 32

test_diff = 12

alpha = 0.01 / 2

# get critical value
crit_val = abs(round(st.norm.ppf(alpha), 4))
print(f"Critical value: {crit_val}")

# calculate difference in mean
y_bar_diff = round(y_bar1- y_bar2, 4)
print(f"Difference in means: {y_bar_diff}")

# setup numerator and denomenator
numer = round(y_bar_diff - test_diff, 4)
# print(f"numerator: {numer}")
denom = round(math.sqrt( (var1/n1) + (var2/n2) ), 4)
# print(f"denom: {denom}")
print(numer)
print(denom)

z_val = round(numer / denom, 4)
print(f"Z-value: {z_val}")

p_val = round(1 - st.norm.cdf(z_val), 4)
print(f"P value: {p_val}")
```

```
# 2
alpha = 0.10

# Ho: mu_male - mu_female = 0
# Ha: mu_male - mu_female > 0

nwtq = 20 / 80
dwtq = 8 / 80

p_c = round((20 + 8) / 160, 4)
print(f"Pc: {p_c}")

p_diff = round(nwtq - dwtq, 5)
print(f"P diff: {p_diff}")

crit_val = abs(round(st.norm.ppf(alpha), 4))
print(f"Critical value: {crit_val}")

test_diff = 0

# denom = math.sqrt(p_c* (1 - p_c) * ((1/total_new) + 1/total_old))
# print(denom)

# get test statistic
z_val = round((p_diff - test_diff) / math.sqrt(p_c * (1 - p_c) * ((1/80) + 1/80)), 4)
print(f"Z value: {z_val}")

# get p-value
p_val = round(1 - st.norm.cdf(z_val), 4)
print(f"P value: {p_val}")

# insufficient evidence to conclude males living more than 20 mi away exceeds females living 20 mi away
claim = "there is no difference in proportion among male and female students who live more than 20 miles away from campus"
################################### UPDATE HERE TO DETERMINE CASE 
if z_val > crit_val:
    print(f"At alpha={alpha}, we can conclude there is sufficient evidence to refute \n{claim}")
else:
    print(f"At alpha={alpha}, we can conclude there is not sufficient evidence refute \n{claim}")
```

```
# 3
# small independent sample averages, equal variance

n1 = 11
var1 = 46.11
mean1 = 160.35
df1 = n1 - 1

n2 = 11
var2 = 41.92
mean2 = 143.7
df2 = n2 - 1

mean_diff = mean1 - mean2
var_diff = 16.02  # given


alpha = 0.05
# check variance > F- test
print(f"---- Testing for equal variance ----")
ratio = round(var1 / var2, 4)  # LARGE VALUE GOES ON TOP
print(f"Ratio: {ratio}")

crit_val = round(st.f.ppf(1 - alpha, df1, df2), 4)
print(f"F Critical value: {crit_val}")

f_val = round(var1 / var2, 4)
print(f"F Test statistic: {f_val}\n")

p_val = round(1 - st.f.cdf(f_val,df1,df2), 4)
print(f"P value: {p_val}")

if f_val > crit_val:
    print(f"Assume unequal variance")
    df_num = ((sd1**2 / n1) + (sd2**2 / n2)) ** 2
    df_denom = ( ( (sd1**2 / n1) ** 2) / (n1 - 1) ) + ( ( (sd2**2 / n2) ** 2) / (n2 - 1) )
    # always round down for degrees of freedom
    df = math.floor(df_num / df_denom)
    print(f"Degrees of freedom: {df}")
else:
    print(f"Assume equal variance")
    df = n1 + n2 - 2

# insufficient evidence to conclude unequal variances -> equal variance
print("-------------------------")
print(f"Degrees of freedom: {df}")

# construct 95% CI
alpha = 0.05 / 2

crit_val = round(abs(st.t.ppf(alpha, df)), 4)
print(f"Critical value: {crit_val}")

s_pool = round(math.sqrt( ( ( (n1 - 1) * var1) + ( (n2 - 1) * var2)) / df), 4)
print(f"Pooled variance: {s_pool}")

y_diff = round(mean1 - mean2, 4)
print(f"Mean diff: {y_diff}")

prod = round(crit_val * s_pool * math.sqrt( (1/n1) + (1/n2) ), 4)
print(f"Product: {prod}\n")

lower = round(y_diff - prod, 4)
upper = round(y_diff + prod, 4)

print(f"{round((1 - 2 *alpha) * 100, 0)}% confidence interval: ({lower}, {upper})")
```

```
n1 = 77
n2 = 77

x1 = 35
x2 = 54

p_hat1 = x1 / n1
p_hat2 = x2 / n2
p_diff = round(p_hat1 - p_hat2, 4)
print(f"P diff: {p_diff}")

alpha = 0.10 / 2
crit_val = abs(round(st.norm.ppf(alpha), 4))
print(f"Critical value: {crit_val}")

prod = round(math.sqrt(((p_hat1 * (1 - p_hat1)) / n1) + ((p_hat2 * (1 - p_hat2)) / n2)), 4)
print(f"Multiplier: {round(prod * crit_val, 4)}")

lower = round(p_diff - (crit_val * prod), 4)
upper = round(p_diff + (crit_val * prod), 4)

print(f"{round((1 - (alpha * 2)) * 100, 1)}% confidence interval: ({lower}, {upper})")
```
