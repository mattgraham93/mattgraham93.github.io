```
import math
import scipy.stats as st
import numpy as np
```

### 6.4b 
(equal variances)

```
y_bar1 = 50.3
sd1 = 7.23
var1 = sd1**2
n1 = 13

y_bar2 = 58.6
sd2 = 6.98
var2 = sd2**2
n2 = 21

df = n1 + n2 - 2
print(f"Degrees of freedom: {df}")

alpha = 0.01 / 2  # --> two-tail
crit_val = round(abs(st.t.ppf(alpha, df)), 4)
print(f"Critical value: {crit_val}")

s_pool = round(math.sqrt( ( ( (n1 - 1) * var1) + ( (n2 - 1) * var2)) / df), 4)
print(f"Pooled variance: {s_pool}")

y_diff = round(y_bar1 - y_bar2, 4)
print(f"Mean diff: {y_diff}")

prod = round(crit_val * s_pool * math.sqrt( (1/n1) + (1/n2) ), 4)
print(f"Product: {prod}")

lower = round(y_diff - prod, 4)
upper = round(y_diff + prod, 4)

print(f"{round((1 - (alpha * 2)) * 100, 1)}% confidence interval: ({lower}, {upper})")
```

We are 99% confident that the mean difference between population 1 is 15.1373 to 1.4627 points lower than that of population 2.

### 6.6d 

    Upstream   ave = 4.9067  s = 0.1870        
    Downstream  ave = 3.74    s = 0.3203  
    n = 15
    
(unequal variances) - small sample

```
y_bar1 = 4.9076
sd1 = 0.1870
n1 = 15

y_bar2 = 3.74
sd2 = 0.3203
n2 = 15

df_num = ((sd1**2 / n1) + (sd2**2 / n2)) ** 2
df_denom = ( ( (sd1**2 / n1) ** 2) / (n1 - 1) ) + ( ( (sd2**2 / n2) ** 2) / (n2 - 1) )

# always round down for degrees of freedom
df = math.floor(df_num / df_denom)
print(f"Degrees of freedom: {df}")

alpha = 0.05 / 2

crit_val = round(abs(st.t.ppf(alpha, df)), 4)
print(f"Critical value: {crit_val}")

avg_diff = round(y_bar1 - y_bar2, 4)
prod = round(math.sqrt( (sd1**2 / n1) + (sd2**2 / n2)), 4)
print(f"Multiplier: {round(prod * crit_val, 4)}")

lower = round(avg_diff - (crit_val * prod), 4)
upper = round(avg_diff + (crit_val * prod), 4)

print(f"{round((1 - (alpha * 2)) * 100, 1)}% confidence interval: ({lower}, {upper})")
```

95% confident the average upstream will be 0.9689 and 1.3663 ppm higher than downstream when looking at dissolved oxygen - implying the community dumping sewage into the river is impacting the environment.

```
p_old_tox = p_hat1 = round(old_tox / told_old, 4)
p_new_tox = p_hat2 = round(new_tox / total_new, 4)
print(f"P old tox: {p_old_tox}; P new tox: {p_new_tox}")

p_diff = round(p_hat1 - p_hat2, 4)
print(f"P diff: {p_diff}")

alpha = 0.05 / 2
crit_val = abs(round(st.norm.ppf(alpha), 4))
print(f"Critical value: {crit_val}")

prod = round(math.sqrt(((p_hat1 * (1 - p_hat1)) / n1) + ((p_hat2 * (1 - p_hat2)) / n2)), 4)
print(f"Multiplier: {round(prod * crit_val, 4)}")

lower = round(p_diff - (crit_val * prod), 4)
upper = round(p_diff + (crit_val * prod), 4)

print(f"{round((1 - (alpha * 2)) * 100, 1)}% confidence interval: ({lower}, {upper})")
```

We can conclude that we are 95% confident that there is no difference in proportions of toxic to non-toxic sludge used for agricultural crops when observing the 2 treatment types.
