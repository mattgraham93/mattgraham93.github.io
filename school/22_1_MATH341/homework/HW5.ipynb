#### 5.16 a, b -> z-test

A study is designed to test the hypotheses H0: mu = 26 versus Ha: mu < 26. A random sample of 50 units was selected from a specified population, and the measurements were summarized to y_bar = 25.9 and s = 7.6.

```
# a. With a = .05, is there substantial evidence that the population mean is less than 26?

alpha = 0.05
n = 50
mu = 26

y_bar = x_bar = 25.9
s = 7.6

# calculate critical value
cv = round(st.norm.ppf(alpha), 4)
print(f"Critical value: {cv}")

# calculate test statistic

z_score = round((x_bar - mu) / (s / math.sqrt(n)), 4)
print(f"Z score: {z_score}")

p_val = round(st.norm.cdf(z_score), 4)
print(f"P-value: {p_val}")
```

```
# We can conclude that there is not sufficient evidence to claim that the mean score/value of the population is less than 26.
# When a p-value = 0.463, it is significantly greater than our most-extreme value of 0.05 or less.
```

```
# b.  Calculate the probability of making a Type II error if the actual value of the population mean is at most 24
#       type ii = power = p (z < z(at alpha) + (mu0 - mua) / (sigma/math.sqrt(n)) )

# suppose alpha = 0.05 given passage from a

mu_o = 26
mu_a = 24

# cv carried from part a
# calculate t_val to determine power / probability of type 2 error
z_val = cv - ( (mu_o - mu_a) / (s / math.sqrt(n)) )

print("Type II error: % probability/risk that we are saying the sample mean is invalid when it is valid")
p_tii = st.norm.cdf(z_val)
print(f"P(Type II error when mu = 24) = {round(p_tii, 4)}")
```

#### 5.35 -> t test (one tail)

A study was designed to evaluate whether the population of interest has a mean greater than 9. A random sample of n = 17 units was selected from a population, and the data yield x_bar = 10.1 and s = 3.1.

```
# a. Is there substantial evidence (a = .05) that the population mean is greater than 9?

alpha = 0.05
n = 17
mu = 9
df = n - 1

y_bar = x_bar = 10.1
s = 3.1

# calculate critical value
# we take the absolute value to get the positive t score, as that is our target statistic
cv = abs(round(st.t.ppf(alpha, df), 4))
print(f"Critical value: {cv}")

# calculate test statistic
t_score = round((x_bar - mu) / (s / math.sqrt(n)), 4)
print(f"Z score: {t_score}")

p_val = st.t.cdf(t_score, df)
# return 1 - p_val to get area to the right of the curve
print(f"P-value: {round(1 - p_val, 4)}")

# interpretation done in part b.
```

```
# We can conclude that the sample mean of 10.1 is different from the population. When observing at a confidence level
# of 0.05, we must reject the claim that the mean is greater than 9. However, if we looked at a level of 0.10,
# our interpretation would be different. 
```

```
# b. What is the level of significance of the test
#       The significance level of the test is 0.05, or we are looking at a relative "5%" risk of this specific test
#       returning a sample mean that may fool us into thinking it is valid, when it is an extreme value.
```

#### 10.7b - binomial testing
The test for screening donated blood for the presence of the AIDS virus was developed in the 1980s. It is designed to detect antibodies, substances produced in the body of donors carrying the virus; however, it is not 100% accurate. The developer of the test claimed that the test would produce fewer than 5% false positives and fewer than 1% false negatives. In order to evaluate the accuracy of the test, 1,000 persons known to have AIDS and 10,000 persons known to not have AIDS were given 
the test. The following results were tabulated:

    Test Result     Has Aids    Does Not Have Aids   Total 
    Positive test   993         591                  1,584
    Negative test   7           9,409                9,416
    Total           1,000       10,000               11,000

```
# setup values
# does not have aids
fals_pos = 591
pos_neg = 9409

# has aids
neg_pos = 7
pos_pos = 993

total_no_aids = pos_neg + fals_pos  
total_aids = neg_pos + pos_pos

total = total_aids + total_no_aids
# print(f"{total_aids}, {total_no_aids}, {total}")
```

```
# b. Is there substantial evidence (a = .01) that the test produces less than 5% false positives.
# null: p = 0.05
# alt: p < 0.05

p = 0.05
p_hat = round(fals_pos / total, 4)
print(f"P_hat: {p_hat}")
n = total
alpha = 0.01

# determine test type
np = n * p #  needs to be > 5 for z
nq = n * (1 - p) #  AND needs to be > 5 for z

# print(f"{np}, {nq}")
# since np and nq are both greater than 5, we can use t-tests 
s = math.sqrt( p * (1 - p) / n)

# find critical value
cv = round(st.norm.ppf(alpha), 4)
print(f"Critical value: {cv}")

# calculate test statistic
z_score =  round((p_hat - p) / s, 4)
print(f"Z score: {z_score}")

# calculate p-value
p_val = round(st.norm.cdf(z_score), 4)
print(f"P-value: {p_val}")
```

```
# This feels wrong, but we are looking for < 5% and our sample mean is > 5%...
# We can conclude that there is a significant difference in false positivity rate claims and reality. We can state
# that these tests were highly ineffective when looking across large populations. 
```
