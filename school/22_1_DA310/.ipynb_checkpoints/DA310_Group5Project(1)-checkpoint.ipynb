```python
df = pd.read_csv('https://raw.githubusercontent.com/mattgraham93/mattgraham93.github.io/master/DataSpellProjects/data/insurance.csv')

# check for complete data
print(df.info())
print(df.describe(include = 'all'))
```

#### Implement multiple linear regression
*   Find best independent variables
*   Find interactions between/among independent variables

```python
# get categorical variables
X = df[['age', 'sex', 'bmi', 'children', 'smoker']]
Y = df['charges']
x_ns = train[['age', 'bmi', 'children', 'smoker']]
x_lc = train[['age', 'bmi', 'log_children', 'smoker']]
y_ns = train['charges']
y_lc = train['log_charges']

regr = linear_model.LinearRegression()
regr.fit(X, Y)
print('Intercept: \n', regr.intercept_)
print('Coefficients: \n', regr.coef_)

# with statsmodels
X = sm.add_constant(X) # adding a constant
x_ns = sm.add_constant(x_ns)

ns_model = sm.OLS(y_ns, x_ns).fit()
lc_model = sm.OLS(y_lc, x_lc).fit()

# how do we print the variables here?

model = sm.OLS(Y, X).fit()
predictions = model.predict(X)
```

#### Predict the test dataset and get the performance

```python
#in progress
#testX = test[['age', 'sex', 'bmi', 'children', 'smoker']]
testY = test['log_charges']
#testX = sm.add_constant(testX)
#predY = model.predict(testX)
#print(chi2(testY, predY))
#print(predY)
print(testY)

#testX_ns = test[['age', 'bmi', 'children', 'smoker']]
#testX_ns = sm.add_constant(testX_ns)
x_lc = test[['age', 'bmi', 'log_children', 'smoker']]
predY_ns = lc_model.predict(x_lc)
print(st.chisquare(testY, predY_ns))
print(predY_ns)
print('Average Difference: ', sum(predY_ns.values - testY.values)/len(testY))
```

```python
#sns.histplot(testY,color = "lightblue", label="Test")
#sns.histplot(predY_ns, color = "lightpink", label="Predict")
#plt.legend()
#plt.show()

#create the overlapping histograms here:
plt.hist(testY, color='blue', label='Test', alpha=0.5)
plt.hist(predY_ns, color='green', label='Predict', alpha=0.5)
plt.legend()
plt.show()
```

```python
plt.bar(testY, predY_ns)
plt.show()
```

```python
# Goodness of fit test
k = len(test)
df_k = k - 1
alpha = 0.001

crit_val = round(st.chi2.ppf(1 - alpha, df_k), 4)
print(f"Critical value: {crit_val}")

chi_sq_calc = sum( ( (test.charges.values - predY_ns.values)**2) / predY_ns.values)
print(f"Test statistic: {round(chi_sq_calc, 4)}")

claim = "our model is a good fit"

if chi_sq_calc < crit_val:
    print(f"At alpha = {alpha}, there is sufficient evidence to state {claim}")
else:
    print(f"At alpha = {alpha}, there is not sufficient evidence to state {claim}")

print(f"P-value: {st.chi2.cdf(1 - chi_sq_calc, df_k)}")
```
