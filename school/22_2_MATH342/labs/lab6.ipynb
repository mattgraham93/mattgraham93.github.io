```python
import pandas as pd
import scipy.stats as st
```

```python
# determining slope equivalence

slope_setup = {
    'treatment': ['C', 'I', 'II'],
    'slope': [3.910, 3.252, 3.739],
    'sse': []
}

slopes = pd.DataFrame(slope_setup)
slopes.set_index('treatment', inplace=True)
slopes
```

```python
sse_f = 486.9
sse_ii = 614
N = 30
t = 3

df_1 = df_t = t - 1
df_2 = N - (2 * t)

print(f"df1:{df_1}\ndf2:{df_2}")
```

```python
numer = (sse_ii - sse_f) / (t - 1)
denom = sse_f / (N - (2 * t))

f_stat = numer / denom
print(f"F statistic: {f_stat}")

p_val = st.f.cdf(f_stat, df_1, df_2)
print(f"P val: {1 - p_val}")
```

```python
alpha = 0.05

f_crit = st.f.ppf(1 - alpha, df_1, df_2)
print(f"Critical value: {f_crit}")
```
