#### a
- b0 - b3 are mathematical constants, scalars
-- b0: default, not in program, english not spoken at home
-- b1: in program, english not spoken
-- b2: not in program, english spoken
-- b2: in program, english spoken

- b2 * E means it is true that the student grew up in a home where English was spoken - this will help tell us whether or not English being spoken at home had an impact on their written scores

- Standard error is given, or assumed 0 if variance is normally distributed

#### c - 95% confidence interval

```python
b_dist = 0.169
se_dist = 0.152

df_err = 19

t = abs(st.t.ppf(0.05 / 2, df_err))
me_dist = t * se_dist

lower = b_dist - me_dist
upper = b_dist + me_dist
print(f"95% confidence interval: ({round(lower, 4)}, {round(upper, 4)})")
```

We can note that we are 95% confident the amount distance impacts the predicted value ranges from -.1491 and .4871. Through this range, we can see why distance does not provide value to our model's predicted revenue.

#### d - hypothesis testing

b2 = slope associated with population

Ho: b2 = 0.5
Ha: b2 > 0.5

alpha = 0.05
n = 22

```python
b_pop = 1.094
b2 = 0.5
se_pop = 0.202

t = (b_pop - b2) / se_pop
n = 22
df = n - 1
print(f"Test statistic: {round(t, 4)}")

alpha = 0.05
crit_val = abs(st.t.ppf(alpha, df))
print(f"Critical value: {round(crit_val, 4)}")

p_val = st.t.cdf(t, df)
print(f"P value: {round(1 - p_val, 4)}")
```

When observing the slope of our linear regression model, we can state we are 95%, and even 99% confident that our slope associated with population is greater than 0.5.
